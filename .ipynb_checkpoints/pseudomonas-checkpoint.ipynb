```python
import pandas as pd
import numpy as np

from sklearn import preprocessing

from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers import BatchNormalization
from keras.optimizers import RMSprop, SGD
from keras import backend as K
from keras.regularizers import WeightRegularizer, l1, l2, activity_l2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw = pd.DataFrame.from_csv('data/pseudomonas.tsv', sep='\t').transpose()
```

```python
y = np.array(raw)
for i in range(raw.shape[0]):
    y[i,:] = y[i,:] / np.sum(y[i,:])
x = preprocessing.scale(y)
```

```python
ae = Sequential()
ae.add(Dense(50, input_dim=x.shape[1],W_regularizer=l2(.000001), activity_regularizer = activity_l2(.000001)))
ae.add(Activation('relu'))
ae.add(Dense(x.shape[1], W_regularizer=l1(0.0005)))
ae.add(Activation('softmax'))
ae.compile(optimizer=SGD(lr=.01, momentum = 0.9, Nesterov = True), loss='categorical_crossentropy')
```

```python
history = ae.fit(x, y, batch_size=64, nb_epoch=100, verbose=1)
```

```python
ae.summary()
```

```python
get_code = K.function([ae.layers[0].input], [ae.layers[1].output])
code = get_code([x])[0]
pd.DataFrame(code).describe()
```

```python
weights = ae.get_weights()[2].flatten()
plt.hist(weights, normed=False, bins=50)
plt.ylabel('counts');
```

```python
import palettable as pal
cmap=pal.colorbrewer.diverging.PRGn_11.get_mpl_colormap()

df=pd.DataFrame(ae.get_weights()[2])
```

```python
fig, ax = plt.subplots(1)
fig.set_size_inches(12,8)
h=ax.pcolor(df, cmap=cmap)
ax.margins(0,0)
fig.colorbar(h)
```
